### **Import Data**

### **Split Data to train and test**

```python
# x(input) y(output)

x = df.drop(columns = 'Outcome' )
y = df['Outcome']
```

```python
x
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=1, stratify=y)
```

```python
X_train.shape #614,8
y_train.shape #614,1
X_test.shape #154, 8
y_test.shape #154,1
```

### **Choosing Model**

```python
svm = svm.SVC(kernel='linear')
svm.fit(X_train, y_train)
```

### **Evaluating Model**

```python
#Checking Prediction on test data
X_test_pred = svm.predict(X_test)
test_data_accuracy = accuracy_score(X_test_pred, y_test)
print(test_data_accuracy)
```

```python
#Checking Prediction on train data
X_train_pred = svm.predict(X_train)
train_data_accuracy = accuracy_score(X_train_pred, y_train)
print(train_data_accuracy)
```

### **Making Predictive system**

```python
# input_data = (6,148,72,35,0,33.6,0.627,50)

input_data = (
    0,          # Pregnancies
    140,        # Glucose (high glucose level, common in diabetics)
    82,         # Blood Pressure
    35,         # Skin Thickness
    150,        # Insulin (high, suggests insulin resistance)
    33.6,       # BMI (overweight / obese range)
    0.672,      # Diabetes Pedigree Function (higher genetic risk)
    45          # Age (older age, higher risk)
)

#Converting input data into array
input_data_as_numpy_array = np.asarray(input_data)

#Reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

actual_prediction = svm.predict(input_data_reshaped)
print(actual_prediction)

if (actual_prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
