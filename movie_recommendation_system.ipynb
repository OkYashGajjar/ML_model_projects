# Import necessary dependencies

## USE : Cosine Similarity Algorithm

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

# Data collection

```python
movies = pd.read_csv('/content/movies.csv')
movies
```

# Data PreProcessing

```python
movies.shape
```

```python
movies.info()
```

```python
#Hence it is content based movie recommandation system
movies['release_date'] = pd.to_datetime(movies['release_date'])
movies['release_year'] = movies['release_date'].dt.year
Selection_columns = ['original_title','overview','genres','keywords','director','cast','release_year' ]
Selection_columns
```

```python
for column in Selection_columns :
  movies[column] = movies[column].fillna('')
```

```python
movies.info()
```

```python
combined_features = (
    movies['original_title'] + ' ' +
    movies['overview'] + ' ' +
    movies['genres'] + ' ' +
    movies['keywords'] + ' ' +
    movies['director'] + ' ' +
    movies['cast'] + ' ' +
    movies['release_year'].astype(str)
)
combined_features
```

# Feature Extraction

```python
feature_extraction = TfidfVectorizer()
feature_vectors = feature_extraction.fit_transform(combined_features)
print(feature_vectors)
```

# Getting similarity score with cosine similarity

```python
#Similarity Score
similarity = cosine_similarity(feature_vectors)
print(similarity)
```

```python
similarity.shape
```

```python
#List of movies
movies_list = movies['original_title'].tolist()
movies_list
```

# Final Movie Recommandation System

```python
movie_input = input('Enter your movie name : ')

find_closest_match = difflib.get_close_matches(movie_input, movies_list)
if find_closest_match :
    close_match = find_closest_match[0]

    # finding the index of movie
    index_close_match_movie = movies[movies['original_title'] == close_match]['index'].values[0]

    #Getting List of similar movies
    similarity_score = list(enumerate(similarity[index_close_match_movie]))

    sorted_similar_movies = sorted(similarity_score, key= lambda x:x[1], reverse=True)
    final_similar_movies = sorted_similar_movies[0:21]

    #Suggestion of movies

    print('Suggested Movies : \n')

    i=1

    for movie in final_similar_movies :
      index = movie[0]
      title_from_index = movies[movies.index == index]['original_title'].values[0]
      print(title_from_index)
else :
  print('Movie not found')
```
