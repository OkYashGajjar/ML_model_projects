### **Separate Data (X,Y)**

```python
x = df.drop('price', axis=1)
y = df['price']
```

```python
y
```

```python
realx = x.drop(['stories','guestroom','hotwaterheating','mainroad','basement','airconditioning','prefarea','furnishingstatus'], axis=1)
print(realx)
```

### **Splitting training and testing data**

```python
X_train,X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
print("Train Shapes  : " , X_train.shape, y_train.shape)
print("Test shapes : ", X_test.shape, y_test.shape)
```

```python
categorical_columns = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus']
```

```python
X_train_encoded = pd.get_dummies(X_train, columns=categorical_columns, drop_first=True)
X_test_encoded = pd.get_dummies(X_test, columns=categorical_columns, drop_first=True)
```

### **Choosing Model**

```python
rf = RandomForestRegressor()
rf.fit(X_train_encoded, y_train)
```

```python
x_train_rf_pred = rf.predict(X_train_encoded)
x_test_rf_pred = rf.predict(X_test_encoded)
print(x_train_rf_pred)
print(x_test_rf_pred)
```

```python
#model evaluation

train_rf_MSE = mean_squared_error(x_train_rf_pred,y_train)
print("Train MSE : ", train_rf_MSE)
test_rf_MSE = mean_squared_error(x_test_rf_pred,y_test)
print("Test MSE : ", test_rf_MSE)

train_rf_r2 = r2(x_train_rf_pred,y_train)
test_rf_r2 = r2(x_test_rf_pred,y_test)
print("Train R2 : ", train_rf_r2)
print("Test R2 : ", test_rf_r2)
```

### **Input Prediction**

```python
input_data = (7420, 4, 2, 3, 'yes', 'no', 'no', 'no', 'yes', 2, 'yes', 'furnished')

# Manual mapping
binary_map = {'yes': 1, 'no': 0}
input_data_encoded = [
    input_data[0],  # area
    input_data[1],  # bedrooms
    input_data[2],  # bathrooms
    input_data[3],  # stories
    binary_map[input_data[4]],  # mainroad
    binary_map[input_data[5]],  # guestroom
    binary_map[input_data[6]],  # basement
    binary_map[input_data[7]],  # hotwaterheating
    binary_map[input_data[8]],  # airconditioning
    input_data[9],              # parking
    binary_map[input_data[10]], # prefarea
]

# One-hot encode 'furnishingstatus' if needed
furnishing_status_map = {
    'unfurnished': [1,0],
    'semi-furnished': [0,1],
    'furnished': [0,0]
}
furnishing_onehot = furnishing_status_map[input_data[11]]

# Combine everything
final_input = input_data_encoded + furnishing_onehot

# Convert to numpy array
input_data_array = np.asarray(final_input).reshape(1, -1)

# Predict
prediction = rf.predict(input_data_array)
print("Predicted price:", prediction)
```
