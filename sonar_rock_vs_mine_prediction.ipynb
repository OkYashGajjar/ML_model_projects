### **Import Data**



```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('Copy of sonar data.csv', header=None)
df.head()
```

### **Data Pre-Processing**

```python
df.head()
```

```python
df.shape # 208 Rows and 6 column
df.info() # (all columns are non null and float64 except last one is object)
```

```python
# Examples of rock and mine
df[60].value_counts() #Metal 111 rows, Rock 97 rows
```

```python
df.groupby(df[60]).mean()
```

### **Separating Data and Label**

```python
# x(Data) , y=(label)

x = df.drop(columns=60, axis=1)
y = df[60]
x.shape #208 Rows and 60 Columns total
y.shape #208 Rows and 1(last) Columns total
```

### **Training and Testing data**

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.1,random_state=2) #test_size = 10% = 0.1(for test data)
xtrain.shape #187,60
xtest.shape #21,60
ytrain.shape #187,1
ytest.shape #21,1
```

### **Choosing Model**

```python
lr = LogisticRegression()
lr.fit(xtrain,ytrain)
```

### **Evaluating Data**



```python
#for evaluation we will check on train data
xtrain_pred = lr.predict(xtrain)
#Checking Accuracy of xtrain_pred data over ytrain (it means comparing our predicting value to actual value)
xtrain_acc = accuracy_score(xtrain_pred, ytrain)
print(xtrain_acc) #0.82 is pretty good
```

```python
#Accuracy  Score of test data

xtest_pred = lr.predict(xtest)
xtest_acc = accuracy_score(xtest_pred,ytest)
print(xtest_acc) #0.85 is pretty good
```
