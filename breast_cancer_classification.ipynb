# Let's breakdown about **Tumor**
1.   Bening Tumor
2.   Maligant Tumor

*   **Bening Tumor** : non-cancerous, capsulized, non-invensive, slow-growing, cells are normal
*   **Maligant Tumor** : cancerous, non-capsulized, fast-growing, spread to other parts of body, cells have abnormal shape





## import dependancies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## Load Data

```python
## Load Data
tumor_data = pd.read_csv('/content/data.csv')
tumor_data
```

## Data Pre-Processing

```python
tumor_data.info()
```

```python
tumor_data = tumor_data.drop(['Unnamed: 32','id'], axis=1)
```

```python
#Checking null values
tumor_data.isnull().sum()
```

```python
#No null values are there
tumor_data.describe()
```

```python
tumor_data.shape
```

```python
tumor_data = tumor_data.replace({'diagnosis' : {'M' : 0, 'B' : 1}})
tumor_data.tail()
```

```python
#Checking B & M
tumor_data['diagnosis'].value_counts()
```

```python
tumor_data.groupby('diagnosis').mean()
```

## Separating Feature and Target

```python
# Features = x and target =y  here.

x = tumor_data.drop(columns='diagnosis', axis=1)
y = tumor_data['diagnosis']
```

```python
x.head()
```

```python
y.head()
```

## Splitting data for training and testing

```python
#train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=2)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

## Choosing Model

```python
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
lr = LogisticRegression(max_iter=350, solver='saga')
lr.fit(x_train,y_train)
```

## Testing Prediction Accuracy

```python
# Logistic Regression
```

```python
lr_x_train_pred = lr.predict(x_train)
lr_x_test_pred = lr.predict(x_test)
```

```python
lr_x_train_acc = accuracy_score(y_train, lr_x_train_pred)
lr_x_test_acc = accuracy_score(y_test, lr_x_test_pred)
```

```python
lr_set = pd.DataFrame([{
    'Model': 'Logistic Regression',
    'train_acc': lr_x_train_acc,
    'test_acc': lr_x_test_acc
}])

lr_set
```

```python
#Random Forest
```

```python
rf_x_train_pred = rf.predict(x_train)
rf_x_test_pred = rf.predict(x_test)
```

```python
rf_x_train_acc = accuracy_score(y_train, rf_x_train_pred)
rf_x_test_acc = accuracy_score(y_test, rf_x_test_pred)
```

```python
rf_set = pd.DataFrame([{
    'Model' : 'Random Forest',
    'train_acc' : rf_x_train_acc,
    'test_acc' : rf_x_test_acc
}])
rf_set
```

```python
concate = pd.concat([lr_set, rf_set])
concate.reset_index(drop=True)
```

```python
#Both the models are accuratel, we can use both of them, here Logistic regression has less overfitting so we use that
```
